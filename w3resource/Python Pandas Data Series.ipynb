__10. Write a Pandas program to convert Series of lists to one Series.__

`ds.apply(pd.Series)` - применяет к каждому элементу серии функции создания серии. После чего образуется ДатаФрейм.

`.stack()` - преобразование ДатаФрейма в серию с многоуровневым индексом.

`.reset_index(drop=True)` - сбрасывает индекс. параметр `drop=True` указывает, что индексы будут отброшены, а не добавлены в качестве столбца.

```python
ds = pd.Series([
    ['Red', 'Green', 'White'],
    ['Red', 'Black'],
    ['Yellow']])

print(ds)

print(ds.apply(pd.Series).stack().reset_index(drop = True))
```

__17. Write a Pandas program to get the items which are not common of two given series.__

```python
ds1 = pd.Series([1, 2, 3, 4, 5])
ds2 = pd.Series([2, 4, 6, 8, 10])

ds_union = np.union1d(ds1, ds2) # объединение
ds_intersect = np.intersect1d(ds1,ds2) # пересечение

print(pd.Series(np.setdiff1d(ds_union, ds_intersect))) # вычитание

```

__18. Write a Pandas program to compute the minimum, 25th percentile, median, 75th, and maximum of a given series.__

```python
num_state = np.random.RandomState(100) # создание генератора чисел

ds = pd.Series(num_state.normal(10, 4, 20)) # μ = 10, σ = 4, count = 20

print(ds)

print(np.percentile(ds, q = [0, 25, 50, 75, 100])) 
```

Перцентили - это значения, которые делят данные на части. В данном случае вычисляются следующие перцентили:

- 0-й перцентиль (минимальное значение)
- 25-й перцентиль (значение, ниже которого находится 25% данных)
- 50-й перцентиль (медиана, значение, ниже которого находится 50% данных)
- 75-й перцентиль (значение, ниже которого находится 75% данных)
- 100-й перцентиль (максимальное значение)

__19. Write a Pandas program to calculate the frequency counts of each unique value of a given series.__

```python
ds = pd.Series(np.take(list('0123456789'), np.random.randint(10, size = 40)))
print(ds.head(5))
print(ds.value_counts())
```

`np.take` - выбирает элементы из списка '0123456789' по указанным индексам.
`np.random.randint(10, size=40)` - генерирует 40 случайных чисел из диапазона 0 до 9 (первый параметр 10)

__20. Write a Pandas program to display most frequent value in a given series and replace everything else as 'Other' in the series.__

```python
ds = pd.Series(np.random.randint(1,5, [15]))
print('Оригинальная серия: ')
print(ds)
print('Подсчет элементов в серии:')
print(ds.value_counts())

ds[~ds.isin(ds.value_counts().index[:1])] = 'Other'

print('Итоговая сериая')
print(ds)

```
